```python
# import packages
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# read in dataset using pd.read_csv()
forestFires = pd.read_csv("c:/users/tscott/OneDrive - Eastside Preparatory School/courses/DataScience_2/data/forestfires.csv")
```

```python
# check out contents of the data frame using describe() and head()
forestFires.head()
#forestFires.describe()
```

```python
# create new column for natural log of area, called logArea
forestFires['logArea'] = np.log(forestFires.area + 1)
forestFires.describe()
```

```python
# build feature DF and label DF (one DF of feature columns, and one DF/list of label a.k.a. logArea)
# drop factor columns for month & day
x = forestFires.drop(['area','logArea','month','day'], axis=1)
y = forestFires['logArea']
```

```python
# confirm contents of feature and label data frames
x.head()
#y.head()
```

```python
# (optional) make correlation plot using seaborn's heatmap
correlations = forestFires.corr()
sns.heatmap(correlations, annot=True)
```

```python
# plot each feature column vs logArea using a scatterplot (e.g. seaborn's scatterplot)
# https://stackabuse.com/linear-regression-in-python-with-scikit-learn/
feature_names = x.columns
feature_names

for var in feature_names:
    plt.figure() # Creating a rectangle (figure) for each plot
    # Regression Plot also by default includes
    # best-fitting regression line
    # which can be turned off via `fit_reg=False`
    sns.scatterplot(x=var, y='logArea', data=forestFires).set(title=f'Regression plot of {var} and LogArea');
```

```python
# create train/test split of data, 80/20 split
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2, random_state = 1234)
```

```python
# linear regression
# import from sklearn, build model object, and use fit() to build model
from sklearn.linear_model import LinearRegression

mlr = LinearRegression()  
mlr.fit(x_train, y_train)
```

```python
# output model coefficients
print("Coefficients: ", list(zip(x, mlr.coef_)), "\n Intercept: ", mlr.intercept_)
```

```python
# make predictions based on test data
y_pred = mlr.predict(x_test)
```

```python
# Check model performance with RMSE
from sklearn.metrics import mean_squared_error

# calculate rmse
rmse = np.sqrt(mean_squared_error(y_test, y_pred))
print("The RMSE is: ", rmse)
```
