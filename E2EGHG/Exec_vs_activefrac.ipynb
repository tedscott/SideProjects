```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from pypalettes import load_palette
```

```python
df = pd.read_csv("summary_table.csv")
df
```

```python
# data slices
exec250_comp = df.query('execution_ms == 250.0 & mode == "compute"')
exec250_sleep = df.query('execution_ms == 250.0 & mode == "sleep"')
#exec250_comp
```

```python
plt.figure(figsize=(10,8))
palette = load_palette("AndyWarhol", keep_first_n=3)
sns.boxplot(x="active_fraction", y="std_delta_energy_j", hue="idle_scrapping_ms", data=exec250_comp, palette = palette)
#plt.ylim(0,1)
plt.title("Compute, exec = 250ms")
```

```python
plt.figure(figsize=(10,8))
palette = load_palette("AndyWarhol", keep_first_n=3)
sns.boxplot(x="active_fraction", y="std_delta_energy_j", hue="idle_scrapping_ms", data=exec250_sleep, palette = palette)
#plt.ylim(0,1)
plt.title("Sleep, exec = 250ms")
```

```python
plt.figure(figsize=(10,8))
palette = load_palette("AndyWarhol", keep_first_n=3)
sns.boxplot(x="active_fraction", y="std_delta_energy_j", hue="idle_scrapping_ms", data=df, palette = palette)
plt.ylim(0,3)
plt.title("Compute & sleep, exec = all of them")
```
