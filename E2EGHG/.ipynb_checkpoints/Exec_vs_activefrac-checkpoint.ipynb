## Import raw data to parse for $\Delta E$

```python
df_raw = pd.read_csv("trace_energy_deltas_total.csv")
df_raw1s = df_raw.query('cpu_ms == 1000')
df_raw1s
```

```python
np.unique(df_raw1s['active_interval_ms'])
```

### Filter out active intervals < 25 ms

```python
df_filt_1s = df_raw1s.query('active_interval_ms >= 25.0')
```

```python
plt.figure(figsize=(12,5))

# set seaborn whitegrid theme
sns.set_theme(style="whitegrid")

# Without transparency
sns.kdeplot(
    data=df_filt_1s, x="delta_energy_joules", hue="active_interval_ms", cut=0, fill=False, common_norm=False, alpha=0.5, palette="tab10"
)
plt.show()
```

```python
sns.boxplot(data=df_filt_1s, x="delta_energy_joules", hue="active_interval_ms", palette="tab10")
plt.show()
```

```python
# using small multiple
# create a grid
g = sns.FacetGrid(df_filt_1s, col="active_interval_ms", hue="active_interval_ms", col_wrap=4)

# draw density plots
g = g.map(
    sns.kdeplot, "delta_energy_joules", cut=0, fill=True, common_norm=False, alpha=1, legend=False
)

# control the title of each facet
g = g.set_titles("{col_name}")

# show the graph
plt.show()
```
