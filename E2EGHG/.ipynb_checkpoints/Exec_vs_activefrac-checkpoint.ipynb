## Compute my own summary stats from the trace_energy_totals csv

```python
df_raw
```

```python
df_summary = df_raw.drop(['trace_id','zone','kepler_duration_sum_ms','start_energy_joules',
                         'end_energy_joules','energy_start_ts','energy_end_ts','exper_name',
                         'interval_ms','sleep_ms','idle_scrapping_ms','source_csv'], axis=1)
df_summary
```

```python
df_mean = df_summary.groupby(['cpu_ms','active_interval_ms'], as_index=False).mean()
df_mean=df_mean.rename(columns={'energy_joules':'mean_energy_joules'})
df_std = df_summary.groupby(['cpu_ms','active_interval_ms'], as_index=False).std()
df_std=df_std.rename(columns={'energy_joules':'std_energy_joules'})
#print(df_mean, df_std)
```

```python
# df_temp = df_summary.groupby(['cpu_ms','active_interval_ms'], as_index=False).agg(["mean","std"])
# print(df_temp)
df_summarised = pd.DataFrame([df_mean.cpu_ms,df_mean.active_interval_ms,df_mean.mean_energy_joules,df_std.std_energy_joules]).transpose()
df_summarised
```
